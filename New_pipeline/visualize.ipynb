## Visualize GT

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

def load_boxes(txt_file):
    boxes = []
    with open(txt_file, 'r') as f:
        for line in f:
            parts = line.strip().split()
            class_id,x1, y1, x2, y2 = map(float, parts)
            boxes.append((x1, y1, x2, y2))
    return boxes

def plot_boxes(image_path, txt_path):
    # Load image
    image = Image.open(image_path)
    width, height = image.size

    # Load bounding boxes
    boxes = load_boxes(txt_path)

    # Create plot
    fig, ax = plt.subplots(1)
    ax.imshow(image)

    for x1, y1, x2, y2 in boxes:

        rect = patches.Rectangle(
            (x1, y1), x2 - x1, y2 - y1, 
            linewidth=2, 
            edgecolor='red', 
            facecolor='none'
        )
        ax.add_patch(rect)

    plt.axis('off')
    plt.show()

# Usage example
image_file = "/home/umang.shikarvar/Style_GAN/lucknow/images/8985633_3137890.tif"
txt_file = "/home/umang.shikarvar/Style_GAN/lucknow/labels/8985633_3137890.txt"
plot_boxes(image_file, txt_file)
```
